```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv('/content/titanic.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
```

```python
df.rename(columns={'2urvived': 'survived'}, inplace=True)
```

```python
df = df.loc[:, ~df.columns.str.contains('zero')]
```

```python
print(df.columns)
```

# SKEWNESS

```python
print(df[['age', 'fare']].skew())
```

- Fare is highly positively skewed.
- Age has slight skewness.

```python
df['fare_log'] = np.log1p(df['fare'])

sns.histplot(df['fare_log'], kde=True)
plt.title("Log Transformed Fare")
plt.show()
```

- Log transformation reduces skewness in fare distribution.

# Multicollinearity

```python
corr = df.corr(numeric_only=True)
print(corr)
```

- No very high correlations (>0.8), so multicollinearity is low.
- Pclass and Fare show moderate correlation.

# UNIVARIATE Analysis


```python

plt.figure(figsize=(6,4))
sns.histplot(df['age'], kde=True)
plt.title("Age Distribution")
plt.show()


plt.figure(figsize=(6,4))
sns.histplot(df['fare'], kde=True)
plt.title("Fare Distribution")
plt.show()


plt.figure(figsize=(6,4))
sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.show()
```

Univariate Analysis refers to the analysis of a single variable.
It is used to understand the distribution, central tendency, and spread of the data.

Example: Analyzing the distribution of Age , Fare , Survival using a histogram.

# BIVARIATE ANALYSIS

```python

plt.figure(figsize=(6,4))
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival by Gender")
plt.show()


plt.figure(figsize=(6,4))
sns.countplot(x='pclass', hue='survived', data=df)
plt.title("Survival by Class")
plt.show()


plt.figure(figsize=(6,4))
sns.boxplot(x='survived', y='age', data=df)
plt.title("Age vs Survival")
plt.show()
```

Bivariate Analysis involves the analysis of two variables to understand the relationship between them.

Example: Comparing Survival with Gender to see how gender affects survival rates.

# MULTIVARIATE ANALYSIS

```python

sns.pairplot(df[['survived', 'age', 'fare', 'pclass']])
plt.show()
```

Multivariate Analysis involves analyzing more than two variables simultaneously to understand complex relationships.

Example: Using a pairplot to analyze the relationship between Age, Fare, Passenger Class, and Survival.

#CORRELATION HEATMAP

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

# Final Summary

1. Passenger class had a significant impact on survival, with first-class passengers having the highest survival rates and third-class passengers the lowest.

2. Gender played a crucial role in survival, as female passengers were far more likely to survive compared to male passengers.

3. Age showed a moderate influence on survival, with younger passengers having slightly better survival chances than older ones.

4. Fare was positively correlated with survival, indicating that passengers who paid higher fares (wealthier individuals) had better chances of survival.

5. The dataset revealed clear socio-economic disparities, where wealth and class influenced survival outcomes.

6. The fare feature showed strong positive skewness, which was handled using log transformation to normalize the distribution.

7. Correlation analysis indicated that there was no strong multicollinearity among features, though passenger class and fare showed moderate correlation.

8. Overall, the analysis highlighted that survival was not random but influenced by multiple factors such as class, gender, and economic status.
